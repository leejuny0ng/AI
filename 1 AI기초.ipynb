```python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(example)

result = []

for word in word_tokens:
  if word not in stop_words: # 불용어 아닌것을 result 배열에 추가
    result.append(word)

print("불용어 제거 전:", word_tokens)
print("불용어 제거 후:", result)
```

```python
okt = Okt()

example = "고기를 아무렇게나 구우려고 하면 안 돼. 고기라고 다 같은 게 아니거든. 예컨대 삼겹살을 구울 때는 중요한 게 있지."
stop_words = "를 아무렇게나 구 우려 고 안 돼 같은 게 구울 때 는"

stop_words = set(stop_words.split(' '))
word_tokens = okt.morphs(example)

result = [word for word in word_tokens if not word in stop_words]

print('불용어 제거 전 :',word_tokens)
print('불용어 제거 후 :',result)
```

search
- match는 한 단어만 확인하는 반면, search는 문자열의 전체 검사

```python
check = 'ab?'
# ? 앞에 문자가 있는지 또는 없는지를 나타냄
print(re.search("a",check))
# ab에 b가 없으면 a가 되기에 match
print(re.match('kkkab',check))
# kkk는 기존 check에 존재하지 않는 문자이므로 match되지 않음
print(re.search("kkkab", check))
# 상동
print(re.match('ab', check))
# ab에 b가 있는 ab기에 match
```

split
- 정규표현식에 해당하는 문자열을 기준으로 문자열을 나눔


```python
r = re.compile(' ')
print(r.split('abc abbc abcbab'))
# ' ' 공백이 컴파일 되었기에 abc abbc abcbab를 리턴
r = re.compile('c')
print(r.split('abc abbc abcbab'))
# c를 기준으로 compile 된것이기에 ab abb ab bab를 리턴

r = re.compile('[0-9]') # 0부터 9까지 모두 (정수 기준)
print(r.split("slabc 2v3s 4sss 5a"))
# 숫자를 기준으로 comiple 한 것으로  slabc v s sss a 를 리턴
```

sub
- 정규표현식과 일치하는 부분을 다른 문자열로 교체


```python
print(re.sub('[a-z]', '1', 'abcdefg'))
# 영어에 해당된다면 그냥 1로 만들어라
print(re.sub('[^a-z]', '1', 'abc defg'))
# 알파벳에 해당되지 않는다면 1로 만드는 것, 공백은 해당되지 않으니 abc1defg가 나올 것
```

findall
- 컴파일한 정규표현식을 이용해 정규 표현식과 맞는 모든 문자열을 리스트로 변환

```python
print(re.findall('[\d]', '1ab 2cd 3ef 4g'))
# d는 decimal (정수를 뽑음) >> 1, 2, 3, 4를 출력할 것

print(re.findall('[\W]', '!abcd@@#'))
# W는 문자또는 숫자가 아닌 문자(특수문자) >> !, @, @, # 를 출력할 것
```

finditer
- 컴파일한 정규표현식을 이용해 정규 표현식과 맞는 모든 문자(열)을 `iterator` 객체로 반환
- `iterator` 객체를 이용하면 생성된 객체를 하나씩 자동으로 가져올 수 있어 처리가 간편

```python
iter1 = re.finditer('[\d]', '1ab 2cd 3ef 4g')
print(iter1)
for i in iter1:
  print(i)

iter2 = re.finditer('[\W]', '!abcd@@#')
print(iter2)
for i in iter2:
  print(i)
```

```python
import re
from konlpy.tag import Mecab

tagger = Mecab()
corpus = "평생 살 것처럼 꿈을 꾸어라. 그리고 내일 죽을 것처럼 오늘을 살아라."
tokens = tagger.morphs(re.sub("(\.)", "", corpus))
# 정규화식을 통해 특수기호를 빼고 replace
vocab = {}
bow = []

for tok in tokens:
  if tok not in vocab.keys():
    vocab[tok] = len(vocab)
    bow.insert(len(vocab) - 1, 1)
  else:
    index = vocab.get(tok)
    bow[index] = bow[index] + 1

print(bow)
print(vocab)
# 일일이 단어 하나까지 뽑아서 비교함
```

어휘 빈도 - 문서 역빈도 분석 (TF-IDF; Term Frequency-Inverse Document Frequency)
- 많이 나온 단어가 그 단어가 특정 문서의 핵심 단어가 아닐 수도 있다는 것
- 단순히 빈도수가 높은 단어가 아닌, 특정 문서에서만 많이 나온 단어라면 핵심 단어로 간주
- TF-IDF = 어휘 빈도 * 역문서 빈도
- 어휘 빈도는 특정 문서에서 특정 단어가 많이 등장하는 것을 의미
$$ tf_{x,y} $$
- 역문서 빈도는 다른 문서에서 등장하지 않는 단어 빈도를 의미
$$ log(N/df_x) $$      
- TF-IDF는 다음과 같이 표현
$$ W_{x,y} = tf_{x,y} * log(N/df_x) $$

- TF-IDF를 편리하게 계산하기 위해  scikit-learn의 tfidfvectorizer를 이용

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words = 'english').fit(corpus)

print(tfidf.transform(corpus).toarray())
print(tfidf.vocabulary_)
```

```python
columns = []
for k,v in sorted(tfidf.vocabulary_.items(), key = lambda item:item[1]):
  columns.append(k)

pd.DataFrame(tfidf.transform(corpus).toarray(), columns = columns)
# 가시성 좋게 DataFrame화
```
